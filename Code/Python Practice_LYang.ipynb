```python
sns.set_theme(style = "darkgrid")

# Histogram
plt.figure(figsize=(8, 5))
plt.hist(df["Pregnancies"], color="steelblue", edgecolor="black", bins=10)
plt.xlabel ("Pregnancies")
plt.ylabel("Count")
plt.title("Histogram of Pregnancies")
plt.show()
```

```python
sns.set_theme(style="darkgrid")

sns.violinplot(df["BMI"], color = "red")
plt.show()
```

```python
# Apply Seaborn styling (gridlines, fonts, spacing)
sns.set_theme(style="whitegrid")  # or "whitegrid", "ticks", etc.

# Plot
plt.figure(figsize=(8, 5))
plt.hist(df["Pregnancies"], 
         color='lightblue', # Change this to your favorite color
         edgecolor='black', 
         bins=16) # Change the bin count to your age
plt.xlabel("Pregnancies")
plt.ylabel("Count")
plt.title("Histogram of Pregnancies")
plt.show()
```
